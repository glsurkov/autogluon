```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
import pathlib

# https://www.kaggle.com/datasets/harshitstark/real-movies-dataset

dataset_path = pathlib.Path('./datasets/movies.csv').resolve()

train_movies_data = TabularDataset('./datasets/movies.csv')

sample_data_train = train_movies_data.sample(350)

print(sample_data_train)

predictor = TabularPredictor(label='Director', problem_type='multiclass').fit(train_data=sample_data_train)
```

```python
sample_test_data = train_movies_data.sample(150)

predictions = predictor.predict(sample_test_data)
print(predictions)
```

```python
# https://www.kaggle.com/competitions/home-data-for-ml-course/data

train_missing_data = TabularDataset('./datasets/missing_values/missing_values_train.csv')

predictor = TabularPredictor(label='SalePrice', problem_type='regression').fit(train_data=train_missing_data)
```

```python
# https://www.kaggle.com/datasets/harshitstark/real-movies-dataset

dataset_path = pathlib.Path('./datasets/movies.csv').resolve()

train_movies_data = TabularDataset('./datasets/movies.csv')

sample_data_train = train_movies_data.sample(350)

model_space = {
    'GBM': [{'extra_trees': True, 'ag_args': {'name_suffix': 'XT'}}, {}, 'GBMLarge'],
    'CAT': {},
}

predictor = TabularPredictor(label='Director', problem_type='multiclass').fit(train_data=sample_data_train, hyperparameters=model_space)
```

```python
hyperparameters = {
    'GBM': [{'extra_trees': True, 'num_boost_round': 50,'early_stopping_rounds': 5, 'ag_args': {'name_suffix': 'XT'}}, {}, 'GBMLarge']
}

predictor = TabularPredictor(label='Director', problem_type='multiclass', stopping_metric='accuracy').fit(train_data=sample_data_train, time_limit=3600, presets='best_quality', hyperparameters=hyperparameters)
```

```python
from autogluon.vision import ImagePredictor
from torchvision import transforms
from torchvision.datasets import ImageFolder

transform = transforms.Compose([transforms.Resize((32, 32)), transforms.ToTensor()])
train_dataset = ImageFolder(root='./data/cifar10_train', transform=transform)
test_dataset = ImageFolder(root='./data/cifar10_test', transform=transform)

predictor = ImagePredictor()

predictor.fit(train_dataset, early_stop_patience=5, early_stop_min_delta=0.001, early_stop_baseline=0.5, early_stop_max_value=0.95)

results = predictor.evaluate(test_dataset)
print(results)
```
